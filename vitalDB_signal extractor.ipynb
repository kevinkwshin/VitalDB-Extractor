```python
signal_table=[]
table_label=[]
# print(signal_table)
# print(table_label)

for i in range(0,len(device0_signal)+len(device1_signal)+len(device2_signal)):
    if i<len(device0_signal):
        print('i=',i,', device0')
        temp = vr_data.read_wave_datetime_interval(device0_name,device0_signal[i],datetime.datetime(start_year,start_month,start_date,start_hour,start_minute,start_second),datetime.datetime(end_year,end_month,end_date,end_hour,end_minute,end_second))
        signal_table.append(temp[0])
        signal_table.append(temp[1])
        table_label.append("Time")
        table_label.append(device0_signal[i])
        
    elif i<len(device0_signal)+len(device1_signal):
        print('i=',i,', device1')
        temp = vr_data.read_number_datetime_interval(device1_name,device1_signal[i-len(device0_signal)],datetime.datetime(start_year,start_month,start_date,start_hour,start_minute,start_second),datetime.datetime(end_year,end_month,end_date,end_hour,end_minute,end_second))
        signal_table.append(temp[0])
        signal_table.append(temp[1])    
        table_label.append("Time")
        table_label.append(device1_signal[i-len(device0_signal)])
    else:
        print('i=',i,', device2')
        temp = vr_data.read_number_datetime_interval(device2_name,device2_signal[i-len(device0_signal)-len(device1_signal)],datetime.datetime(start_year,start_month,start_date,start_hour,start_minute,start_second),datetime.datetime(end_year,end_month,end_date,end_hour,end_minute,end_second))
        signal_table.append(temp[0])
        signal_table.append(temp[1])
        table_label.append("Time")
        table_label.append(device2_signal[i-len(device0_signal)-len(device1_signal)])
```

```python
from itertools import zip_longest
signal_table_column=list(zip_longest(*signal_table))

csv_file = open(fname_with_time+'.csv', 'w')
with csv_file:
    writer = csv.writer(csv_file)
    writer.writerow(table_label)
    writer.writerows(signal_table_column)

print("Writing complete")
```
